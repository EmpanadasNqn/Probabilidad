### Simulación:
Observar que hay una semilla que permite que sea reproducible, y que las variables son creadas con distribuciones de probabilidad.


```python
import numpy
numpy.random.seed(0)

N = 100

new_data = {}
new_data['gender'] = numpy.random.choice(
    ['Mujer', 'Hombre', 'Otro', 'Prefiero no decirlo'], size=N,
    p=[0.3, 0.65, 0.01, 0.04])
new_data['age'] = numpy.random.normal(35, 15, size=N).astype(numpy.int32)
new_data['zodiac_sign'] = numpy.random.choice(
    ['Aries', 'Tauro', 'Géminis', 'Cáncer', 'Leo', 'Virgo', 'Libra', 'Escorpio',
     'Sagitario', 'Capricornio', 'Acuario', 'Picis'], size=N)
new_data['profession'] = numpy.random.choice(
    ['Programador', 'Profesor', 'Maestro', 'Contador',
     'Ingeniero', 'Economista', 'Estudiante', 'Matemático',
     'Físico'], size=N)
round_100 = lambda x: int(int(x) / 100) * 1000
new_data['junior_programmer_salary'] = numpy.vectorize(round_100)(
    numpy.random.lognormal(mean=8.5, sigma=0.25, size=N))
new_data['senior_programmer_salary'] = numpy.vectorize(round_100)(
    numpy.random.lognormal(mean=9.5, sigma=0.25, size=N))
new_df = pandas.DataFrame(new_data)
new_df
```

```python
new_df[new_df.gender == 'Hombre'].junior_programmer_salary.describe()
```
